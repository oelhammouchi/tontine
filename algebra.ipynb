```python
from sympy import *

n, k, r, t = symbols("n k r t")
w1, w2, w3, w4 = symbols("w1:5")

p_x = Symbol("p_x")
gamma, lamda = symbols("gamma lambda")

C_FIA, Psi_UIA, Psi_UTO, d_TTO = symbols("C_FIA Psi_UIA Psi_UTO d_TTO", cls = Function)

summand = binomial(n - 1, k) * p_x**(k + 1) * p_x**(n - 1 - k) * (1 / (1 - gamma)) * (w1 * C_FIA(t) + w2 * Psi_UIA(t) + w3 * (n * d_TTO(t) / (k + 1)) + w4 * (n * Psi_UTO(t) / (k + 1)))**(1 - gamma)

L = Sum(exp(-r * t) * Sum(summand, (k, 0, n - 1)), (t, 0, oo)) - lamda * (w1 + w2 + w3 + w4 - 1)
```
